```python
from datetime import datetime
class Canteen:
    def __init__(self,user,delivery):
        self.amount=0
        self.valid=True
        self.user=user
        self.delivery=delivery
        self.choice=""
        self.menu={}
        self.order=[]
        try:
            self.canteen_menu()
            self.main()
        except Exception as e:
            print("Error occurred:", e)
       
    def canteen_menu(self):
        try:
            with open("menu.txt", 'r') as file: 
                for line in file:
                    key, item, price = line.strip().split()
                    self.menu[key] = {'item': item, 'price': float(price)}
        except IOError:
            self.valid=False
            print("IO Error: Can't open or read menu file.")
        except ValueError:
            self.valid=False
            print("Value Error: Incorrect format in menu file.")
        
        
    def display_menu(self):
        
        print("\nCanteen Menu:")
        for key, item in self.menu.items():
            print(f"{key}. {item['item']} - {item['price']}Rupis")

    def place_order(self):
        try:
            now = datetime.now()
            with open(self.user.user_name+".txt","a") as file:
                file.write("\nOrder Details : "+now.strftime("%d/%m/%Y %H:%M:%S")+" \n")
        except IOError:
             self.valid=False
             print("IO Error: Can't write to current time in file.")
        
        while True:
            try:
                self.display_menu()
                self.choice =input("Enter the item number to add to your order (press '0' to finish): ")

                if self.choice == '0':
                    break
                elif self.choice in self.menu:
                                        
                    quantity = int(input(f"Enter the quantity for {self.menu[self.choice]['item']}: "))
                    total_price = quantity * self.menu[self.choice]['price']
                    self.order.append({'item': self.menu[self.choice]['item'], 'quantity': quantity, 'total_price': total_price})
                    
                    print(f"{self.menu[self.choice]['item']} (x{quantity}) added to your order.")
                    
                    with open(self.user.user_name+".txt","a") as file:
                        file.write(f" Item : {self.menu[self.choice]['item']} , Quantity : {quantity} , total_price: Rs.{total_price} \n")
                    
                else:
                    print("Invalid item number. Please try again.")
            except ValueError:
                self.valid=False
                print("Value Error: Invalid input for choise or quantity.")
            except IOError:
                self.valid=False
                print("IO Error: Can't write to user's order in file.")
                    
    def view_order_summary(self):
        print("\nOrder Summary:")
        for line in self.order:
            item, quantity, total_price = line.values()
            print(f"{item} (x{quantity}): Rs.{total_price}")

    def view_bill(self,bill_cheak):
        self.total=0
        delivery_charge=0
        
        

        max_item_length = max(len(line["item"]) for line in self.order)
        print("-------------------------------------------------------")
        space=" " * (max_item_length)
        print(f"Item{space}Quantity       Total-price")  
        
        for line in self.order:
            item, quantity, total_price = line.values()
            spacing = " " * (max_item_length - len(item) + 4)  
            print(f"{item}{spacing}{quantity}              Rs.{total_price}")  
            self.total += total_price
        
        print()

        if self.delivery:
            if self.total < 400:
                delivery_charge = 5
                print(f"Delivery Charge: Rs.{delivery_charge}")
                self.total += delivery_charge
        
        print("-------------------------------------------------------")
        print(f"Total Amount :  {self.total}")
        self.amount=self.total
        if bill_cheak:
            try:
                with open(self.user.user_name+".txt","a") as file:
                    file.write(f"\nDelivery Charge : Rs.{delivery_charge}\n")
                    file.write(f"Total Amount :  Rs.{self.total}\n")
            except IOError:
                self.valid=False
                print("IO Error: Can't write to user's delivery charge or total in file.")
                    

    def clear_order_history(self):
        try:
            with open(self.user.user_name+".txt","r") as file:
                lines=file.readlines()
                file.close()
        except FileNotFoundError:
            self.valid=False
            print("FileNotFound Error: User's file not found.")
        except IOError:
            self.valid=False
            print("IO Error: Can't read to user's history of order in file.")

        try:
            with open(self.user.user_name+".txt","w") as file:
                file.write(lines[0])
                file.write(lines[1])
                file.write(lines[2])
        except IOError:
            self.valid=False
            print("IO Error: Can't write to file.")

        print("Your Order histroy is cleared !!")
    
        
    def main(self):
        while True:
            print("\nCollege Canteen:")
            print("1. View Menu")
            print("2. Place Order")
            print("3. View Order Summary")
            print("4. View Bill")
            print("5. Clear Order History")
            print("6. Exit")

            choice = input("Enter your choice (1-6): ")

            if choice == '1':
                self.display_menu()
            elif choice == '2':
                self.place_order()
            elif choice == '3':
                if not self.order:
                    print("You Don't Place Any Order !!")
                else:
                    self.view_order_summary()
            elif choice == '4':
                if not self.order:
                    print("You Don't Place Any Order !!")
                else:
                    bill=False
                    self.view_bill(bill)
            elif choice == '5':
                self.clear_order_history()
            elif choice == '6':
                if not self.order:
                    print("You Don't Place Any Order !!")
                    try:
                        with open(self.user.user_name+".txt","a") as file:
                            file.write("You Don't Place Any Order !!")
                    except IOError:
                        print("IO Error: Can't write to user's order file.")
                else:
                    bill=True
                    self.view_bill(bill)
                    payment=Payment(self.total,self.user)
                    payment.choice_payment()
                    if self.delivery:
                        print("Please Wait Your Order will delivar Shortly !!")
                    else:
                        print("Your Order is Being Prepered Shortly !!")
                print("Exiting College Canteen. Goodbye!")
                break
            else:
                print("Invalid choice. Please enter a number between 1 and 6.")
```

```python
from datetime import datetime
class Payment:
    def __init__(self,amount,user):
        self.amount = amount
        self.user=user

    def choice_payment(self):
        check_choice=False
        count_choice=0
        while not check_choice:
            
            if count_choice <2:
                print("Paymet Method : ")
                print("1. Cash Payment")
                print("2. Card Payment")
                print("3. UPI Payment")
                payment_choice=input("Enter Choise : ")
                if payment_choice=='1':
                    check_choice=True
                    case=CashPayment(self.amount,self.user)
                    case.process_payment()
                    try:
                        with open(self.user.user_name+".txt","a") as file:
                            file.write("You have done Payment by Cash")
                    except IOError:
                        print("IO Error: Can't write to user's payment in file.")
        
                elif payment_choice=='2':
                    check_choice=True
                    card=CardPayment(self.amount,self.user)
                    card.process_payment()
                    try:
                        with open(self.user.user_name+".txt","a") as file:
                            file.write("You have done Payment by Card")
                    except IOError:
                        print("IO Error: Can't write to user's payment in file.")
        
                elif payment_choice=='3':
                    check_choice=True
                    upi=UPIPayment(self.amount,self.user)
                    upi.process_payment()
                    try:
                        with open(self.user.user_name+".txt","a") as file:
                            file.write("You have done Payment by UPI")
                    except IOError:
                        print("IO Error: Can't write to user's payment in file.")
        
                else :
                    print("enter vaild Choice.")

                count_choice+=1

            else: 
                check_choice=True
                print("Sorry ! You Can Try Again")
    
    
    

class CashPayment(Payment):
    def __init__(self, amount,user):
        super().__init__(amount,user)
    def process_payment(self):
        print(f"Cash payment of ${self.amount} processed.")


class CardPayment(Payment):
    def __init__(self, amount,user):
        super().__init__(amount,user)
        self.card_number = None
        self.card_holder = None
        self.expiry_date = None
        self.cvv = None

    def validate_card_number(self, card_number):
        if not card_number.isdigit() or len(card_number) != 16:
            return False
        return True

    def validate_expiry_date(self, expiry_date):
        now = datetime.now()
        current_year=now.strftime("%Y")
        if len(expiry_date) != 5 or expiry_date[2] != '/':
            return False
        try:
            month, year = map(int, expiry_date.split('/'))
            if not (1 <= month <= 12 and 0 <= year <= 99 and int(year) > int(current_year[-2:])):
                return False
        except ValueError:
            print("Value Error: Invalid input for choise or quantity.")
            return False
        return True

    def validate_cvv(self, cvv):
        if not cvv.isdigit() or len(cvv) != 3:
            return False
        return True


    def process_payment(self):
        self.card_number = input("Enter card number: ")
        while not self.validate_card_number(self.card_number):
            print("Invalid card number.")
            self.card_number = input("Enter card number: ")

        self.card_holder = input("Enter card holder's name: ")

        self.expiry_date = input("Enter expiry date (MM/YY): ")
        while not self.validate_expiry_date(self.expiry_date):
            print("Invalid expiry date.")
            self.expiry_date = input("Enter expiry date (MM/YY): ")

        self.cvv = input("Enter CVV: ")
        while not self.validate_cvv(self.cvv):
            print("Invalid CVV.")
            self.cvv = input("Enter CVV: ")
        print(f"Card payment of ${self.amount} processed using card ending in {self.card_number[-4:]}.")


class UPIPayment(Payment):
    def __init__(self, amount,user):
        super().__init__(amount,user)
        self.upi_id = None
        self.pin = None

    def validate_upi_id(self, upi_id):
        if '@' not in upi_id:
            return False
        return True

    def validate_pin(self, pin):
        if not pin.isdigit() or len(pin) != 4:
            return False
        return True

    def process_payment(self):
        self.upi_id = input("Enter UPI ID: ")
        while not self.validate_upi_id(self.upi_id):
            print("Invalid UPI ID.")
            self.upi_id = input("Enter UPI ID: ")

        self.pin = input("Enter Your PIN: ")
        while not self.validate_pin(self.pin):
            print("Invalid PIN.")
            self.pin = input("Enter Your PIN: ")

        print(f"Mobile payment of ${self.amount} processed to {self.upi_id}.")
```
